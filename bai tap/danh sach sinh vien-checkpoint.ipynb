```python
dic={}
st_l = []
lit = []
class Student():
    def __init__(self,ms,ht,date,qq,cn,lop):
        self.ma_so = ms
        self.ho_ten = ht
        self.date = date
        self.que_quan = qq
        self.nganh = cn
        self.lop = lop
    def check(self):
        print('Mã số: ',self.ma_so)
        print('Họ tên: ',self.ho_ten)
        print('Ngày sinh: ',self.date)
        print('Quê quán: ',self.que_quan)
        print('Chuyên ngành: ',self.nganh)
        print('Lớp: ',self.lop)
        
        
while True:
    print('Nhập 1 để xem danh mục sinh viên')
    print('Nhập 2 để thêm sinh viên mới vào danh mục')
    print('Nhập 3 để Cập nhật thông tin của sinh viên trong danh mục')
    print('Nhập 4 để Xóa thông tin sinh viên khỏi danh mục')
    print('Nhập 5 để Tìm kiếm thông tin sinh viên trong danh mục theo mã số')
    print('Nhập 0 để thoát')
    a = int(input())
    if a==1:
        if st_l==[]:
            print('Danh sách trống !')
        else:
            for i in st_l:
                i.check()
                print('')
    elif a==2:
        ms = str(input('nhập mã số'))
        ht = str(input('mời nhập họ và tên'))
        date = str(input('mời nhập ngày tháng năm sinh,yyyy/mm/dd'))
        qq = str(input('mời bạn nhập quê quán'))
        cn = str(input('mời bạn nhập chuyên ngành'))
        lop = str(input('mời bạn nhập tên lớp'))
        lit.append(ms)
        ms = Student(ms,ht,date,qq,cn,lop)
        st_l.append(ms)
    elif a==3:
        cap_nhat = str(input('mời bạn nhập mã số sinh viên cần cập nhật: '))
        for i in lit:
            if i == cap_nhat:
                print('mời bạn nhập các thông tin cập nhật')
                ht = str(input('mời nhập họ và tên'))
                date = str(input('mời nhập ngày tháng năm sinh,yyyy/mm/dd'))
                qq = str(input('mời bạn nhập quê quán'))
                cn = str(input('mời bạn nhập chuyên ngành'))
                lop = str(input('mời bạn nhập tên lớp'))
                for i in st_l:
                    if i.ma_so == cap_nhat:
                        st_l.remove(i)
                        i = Student(cap_nhat,ht,date,qq,cn,lop)
                        st_l.append(i)
    elif a==4:
        xoa = str(input('mời nhập mã số sinh viên cần xóa: '))
        for i in st_l:
            if i.ma_so == xoa:
                print('đã xóa sinh viên',i.ho_ten,'mã số:',xoa)
                st_l.remove(i)
    elif a==5:
        t = str(input('mời nhập mã số sinh viên để tra cứu thông tin'))
        for i in st_l:
            if i.ma_so == t:
                i.check()
    elif a==0:
        break
print('cám ơn bạn đã sử dụng dịch vụ!')
```
